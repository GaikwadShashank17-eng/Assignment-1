```python

lst = [1, 2, 3, 4, 5]

# Indexing
print(lst[2])      # 3

# Negative indexing
print(lst[-1])     # 5

# Slicing
print(lst[1:4])    # [2, 3, 4]
```

```python
lst = [10, 20, 30, 40]

# Update
lst[1] = 25

# Append
lst.append(50)

# Insert
lst.insert(2, 15)

# Delete
lst.remove(30)

print(lst)
```
